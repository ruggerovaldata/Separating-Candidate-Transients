```python
import pandas as pd
from astropy import units as u
from astropy.coordinates import SkyCoord
import numpy as np
```

```python
data = pd.read_csv('Outliers.csv')
```

```python
runcat, ra, dec = data['runcat'],data['ra'],data['dec']
```

```python
results = []
```

```python
for i, val in enumerate(ra):
    coordinates = SkyCoord(ra= val*u.degree, dec=dec[i]*u.degree)
    results.append([runcat[i],coordinates.to_string('hmsdms')])
    print(coordinates)
```

```python
for i, val in enumerate(ra):
    coordinates = SkyCoord(ra= val*u.degree, dec=dec[i]*u.degree)
    print(runcat[i],coordinates.ra.degree, coordinates.dec.degree)
    results.append([runcat[i],coordinates.to_string('hmsdms')])
```

```python
results = np.array(results)
final_results = []
#print(results) 
for val in results: 
    line = val[1]
    #print(line)
    line = line.split()
    temp=[]
    temp.append(val[0])
    for coord in line: 
        #print(coord)
        coord = coord.replace('h',':')
        coord = coord.replace('m',':')
        coord = coord.replace('s','')
        coord = coord.replace('d',':')
        coord = coord.replace('m',':')
        coord = coord.replace('s','')
        temp.append(coord)
    final_results.append(temp)

print(final_results)
```
